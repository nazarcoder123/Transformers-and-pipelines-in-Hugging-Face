## zero-shot-classification

```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification")
classifier(
     "This is a course about the transformers library",
      candidate_labels=["education","politics","business"]
)
```

```python
classifier = pipeline("zero-shot-classification")
result1 = classifier(
    "This is a game studio where every one come and play game based on regular basic",
candidate_labels=["fun","joy","happines"])

result2 = classifier(
    "There is no man his name is there is no man",
candidate_labels=["idot","genius","mad"]
)

print(result1)
print(result2)
```

## Fill Mask

```python
from transformers import pipeline
```

```python
unmasker = pipeline("fill-mask")
unmasker("This course will teach you all <mask> models.",top_k=2)
```

## Entity recognization

```python
from transformers import pipeline
```

```python
ner = pipeline("ner",grouped_entities=True)
ner("My name is sylvain and I work at hugging face in brooklyn.")
```

## The question answering

```python
from transformers import pipeline
```

```python
question_answerer = pipeline("question-answering")
question_answerer(
   question="Where do I work?",
   context="My name is Nazar and I work at hugging Face in Brookly" 
)
```

```python
summarizer = pipeline("summarization")
summarizer("""Im Nazar born in gulbarga in karnataka completed bachelor of engineer in Data scienist""")
```
